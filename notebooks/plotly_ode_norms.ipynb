```python
import plotly.graph_objects as go
```

```python
import numpy as np
from scipy.stats import norm
import torch
```

```python
steps = [100, 200, 1000]
```

```python
vectors = [
    torch.load(f'latent_ode_encs_{s}.pth').numpy() for s in steps
]
vectors = np.array(vectors)
vectors.shape
```

```python
IDX = 3
```

```python
bins, counts = [], []
for vv in vectors:
    vv = vv.reshape(-1, 768)[:, IDX]
    print(vv.shape)
    c, b = np.histogram(vv, bins=np.linspace(-20, 20, 100), density=True)
    bins += [b]
    counts += [c]
```

```python
fig = go.Figure()
for b, c, s in zip(bins, counts, steps):
    fig.add_trace(go.Scatter(x=b, y=c,
                        mode='lines',
                        name=f'steps = {s}'))
    
probs = norm.pdf(bins[0], 0, 1)

fig.add_trace(go.Scatter(x=bins[0], y=probs,
                    mode='lines',
                    name=f'true st. gauss'))
fig.update_layout(
    title="Распределение в момент времени 1, если шагать от 0 по ODE",
    xaxis_title="x",
    yaxis_title="p(x)",
)

fig.update_xaxes(range=[-20, 20])
fig.show()
```
