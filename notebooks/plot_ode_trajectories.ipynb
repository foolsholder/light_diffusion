```python
import plotly.graph_objects as go
import numpy as np
from scipy.stats import norm
import torch
```

```python
steps = [100, 200, 500]
```

```python
vectors = [
    torch.load(f'tr_{s}.pth').numpy() for s in steps
]
for vv in vectors:
    print(vv.shape)
```

```python
b_i = 4
pos_i = 10
feat_i = 3
```

```python
fig = go.Figure()
for vv, s in zip(vectors, steps):
    timesteps = np.linspace(1/s, 1, s)
    fig.add_trace(go.Scatter(x=timesteps, y=vv[:, b_i, pos_i, feat_i],
                        mode='lines',
                        name=f'steps = {s}'))
fig.update_layout(
    title=f"Trajectory of one particle x_[{b_i}][{pos_i}][{feat_i}]",
    xaxis_title="t",
    yaxis_title="x(t)",
)

fig.update_xaxes(range=[0, 1])
fig.show()
```

```python
fig.write_image("trj_fig.png")
```
